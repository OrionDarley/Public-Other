```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.max_columns = None
pd.set_option('display.max_rows', None)
pd.options.display.max_rows = None
```

```python
deaths_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv'
confirmed_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'

dea = pd.read_csv(deaths_url)
con = pd.read_csv(confirmed_url)

cols = ['iso2', 'iso3', 'code3', 'Combined_Key', 'Country_Region', 'UID', 'FIPS']

dea = dea.drop(cols, axis = 1) #dont drop 'Province_State', will need for Tableau
con = con.drop(cols, axis = 1) #dont drop 'Province_State', will need for Tableau

dea = dea[(dea['Province_State'] == 'Texas')]
con = con[(con['Province_State'] == 'Texas')]

con.tail()
```

```python
con.tail()
```

#### Confirmed Cases Data Prep

```python
date_columns = con.iloc[:, 4:].columns
date_columns
```

```python
con2 = pd.pivot_table(con, index=['Admin2'], values=date_columns, aggfunc=np.sum)
con2
```

```python
con2 = con2.T

date_columns
```

```python
bigdf = pd.DataFrame()

for c in con2.columns:
    df = pd.DataFrame(con2[c]).rename(columns={c: 'confirmed'})
    df['Admin2'] = c
    #After Transposing Admin2 is named the index, remove this
    df['Time Stamp'] = df.index 
    df = df.set_index(df['Time Stamp'])
    
    bigdf = bigdf.append(df)
    
bigdf_con = pd.merge(bigdf, con[['Lat', 'Long_', 'Admin2']], on = 'Admin2', how = 'left')
bigdf_con.tail()
```

#### Death Cases Data Prep

```python
date_columns = dea.iloc[:, 5:].columns

dea2 = pd.pivot_table(dea, index=['Admin2'], values=date_columns, aggfunc=np.sum)

dea2 = dea2.T

bigdf = pd.DataFrame()

for c in dea2.columns:
    df = pd.DataFrame(dea2[c]).rename(columns={c: 'deaths'})
    df['Admin2'] = c
    #After Transposing Admin2 is named the index, remove this
    df['Time Stamp'] = df.index
    df = df.set_index(df['Time Stamp'])
    
    bigdf = bigdf.append(df)
    
bigdf_dea = pd.merge(bigdf, dea[['Lat', 'Long_', 'Admin2', 'Population']], on = 'Admin2', how = 'left')
bigdf_dea.tail()
```

```python
bigdf2 = pd.merge(bigdf_con, bigdf_dea[['Admin2', 'Time Stamp', 'deaths','Population']], 
                  on = ['Admin2', 'Time Stamp'], how = 'left')
bigdf2['State'] = str('Texas')
bigdf2 = bigdf2.rename(columns={'confirmed':'confirmed_cumsum', 'deaths': 'deaths_cumsum'})
```

```python
bigdf2.to_csv('jh_transformed.csv')
```
